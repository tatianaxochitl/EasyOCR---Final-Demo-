```python
def find_episode_links(soup): 
  titles = soup.find_all('td', class_='title')
  episode_links = []
  for title in titles: 
    episode_links.append(title.a['href'])
  return episode_links
```

```python
def find_image_links(episode_url):
  img_data = requests.get(episode_url).text
  img_soup = BeautifulSoup(img_data)
  img_links = []
  images = img_soup.find_all('img', alt="comic content")
  for image in images: 
    img_links.append(image['src'])
  return img_links
```

```python
def create_corpus(url,filename):
  the_file = open(filename,"a")
  data = requests.get(url).text
  soup = BeautifulSoup(data)
  all_eps = find_episode_links(soup)   
  for ep in all_eps: 
    ep_link = 'https://comic.naver.com' + ep
    all_images = find_image_links(ep_link)
    episode_text = []
    for img in all_images[1:]: 
      urllib.request.urlretrieve(img, "sample.jpg")
      img_text = reader.readtext('sample.jpg', detail = 0)
      for line in img_text:
        the_file.write(line)
  return
```

```python
#Ok tokenizing using Open Korean Text through KoNLPy 
# creating a version that stems the input and one that does not
from konlpy.tag import Okt
okt = Okt()
non_stem_corpus = okt.pos(comic_corpus, norm=True)
stem_corpus = okt.pos(comic_corpus, norm=True, stem=True)
```

```python
non_stem_corpus[:10]
```

```python
stem_corpus[:10]
```

Very cool! The stemmed version has the infinitive forms of the verbs and adjectives. Now lets look at some of the words in the corpus. I wanted to see what is most common and in each.

```python
import nltk
res_1 = [(sub[1], sub[0]) for sub in non_stem_corpus]
res_2 = [(sub[1], sub[0]) for sub in stem_corpus]
cfd_non_stem_corpus = nltk.ConditionalFreqDist(res_1)
cfd_stem_corpus = nltk.ConditionalFreqDist(res_2)
```

```python
cfd_non_stem_corpus.conditions()
```

```python
cfd_non_stem_corpus['Noun'].most_common(10)
```

```python
cfd_stem_corpus['Noun'].most_common(10)
```

```python
non_stem_corpus_no_pos = []
stem_corpus_no_pos = []

for a_tuple in non_stem_corpus:
	non_stem_corpus_no_pos.append(a_tuple[0])
 
for a_tuple in stem_corpus:
	stem_corpus_no_pos.append(a_tuple[0])
```

```python
non_stem_corpus_freq = nltk.FreqDist(non_stem_corpus_no_pos)
stem_corpus_freq = nltk.FreqDist(stem_corpus_no_pos)
```

```python
non_stem_corpus_freq.most_common(10)
```

These are all pretty much punctuation or particles put at the end of nouns. 

```python
stem_corpus_freq.most_common(10)
```

These have more verbs so, in stemmed verion to do is the most common word which makes a lot of sense.

```python
a_file = open("/content/drive/MyDrive/stop_words_ko.txt", "r")

stop_words = []
for line in a_file:
  stripped_line = line.strip()
  stop_words.append(stripped_line)

a_file.close()
filtered_non_stem_corpus = [w for w in non_stem_corpus_no_pos if not w in stop_words]
```
